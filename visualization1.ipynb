```python
import matplotlib.pyplot as plt
import numpy as np


fig, ax1 = plt.subplots()
plt.xlabel('Date Time')
plt.ylabel("Sensorx,SensorY,MortorA,MortorB")


plt.title("Please tell it to me the title")

plt.grid(color='black')



ax1.plot(df.DateTimeStamp,df['SensorX'] ,color='red',label='SensorX')
ax1.plot(df.DateTimeStamp,df['SensorY'] ,color='green',label='SensorY')
ax1.plot(df.DateTimeStamp,df['MotorA'] ,color='darkmagenta',label='MotorA')
ax1.plot(df.DateTimeStamp,df['MotorB'] ,color='aqua',label='MotorB')
plt.legend(loc='upper left')

ax2 = ax1.twinx()

ax2.plot(df.DateTimeStamp, df.Temp,color='b',label='Temperature')
plt.ylabel("Temperature")
plt.legend(loc='upper right')


plt.xticks(np.arange(0, 105, 15)) 

plt.rcParams['figure.figsize'] = 15,10

plt.savefig("C:\\Users\\C_v\\Desktop\\chart1.png")


```
